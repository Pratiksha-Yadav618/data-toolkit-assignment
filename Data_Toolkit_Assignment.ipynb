2.How does broadcasting work in NumPy?

--> Broadcasting in NumPy is a powerful mechanism that allows operations between arrays of different shapes without explicitly reshaping them. Instead of writing loops, NumPy automatically “broadcasts” the smaller array across the larger one so that their shapes match.

**Rules of Broadcasting**

1.Compare shapes of the two arrays, starting from the trailing dimensions
(rightmost) and moving left.

2.Two dimensions are compatible if:

They are equal, or

One of them is 1

3.If a dimension is 1 in one array, that array is stretched/repeated along that dimension to match the other array.

21.Why is NumPy's array slicing faster than Python's list slicing?

--> NumPy's array slicing is faster than Python's list slicing because of the way NumPy arrays are stored and how operations are executed.

1.Contiguous Memory Storage

NumPy arrays store data in contiguous blocks of memory (like C arrays).

Python lists store references to objects, which can be scattered in memory.

Accessing contiguous memory is much faster due to CPU caching.

2.Homogeneous Data Type

All elements in a NumPy array are of the same type (integers, floats, etc.).

This allows NumPy to perform vectorized operations and slicing efficiently.

Python lists can store mixed types, so every access requires type checking and pointer dereferencing, slowing it down.

3.Views Instead of Copies

Slicing a NumPy array usually returns a view (a reference to the original data), not a copy.

Python list slicing creates a new list, which takes extra time and memory.

NumPy slicing is significantly faster, especially for large datasets.

4.Vectorized Operations

Slicing in NumPy can be immediately followed by vectorized operations without loops.

Python lists require explicit loops or comprehensions, which are slower in Python.

4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



data = {
    'Age': [23, 34, 29, 40, 35],
    'Salary': [50000, 60000, 55000, 65000, 70000],
    'Experience': [1, 5, 3, 10, 7]
}

df = pd.DataFrame(data)

# Calculate correlation matrix
corr_matrix = df.corr()

print("Correlation Matrix:\n", corr_matrix)

# Visualize with seaborn heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
